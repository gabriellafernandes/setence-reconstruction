```python
# sentence_original = "It is an important step towards equal rights for all passengers"
sentence_original = "Sarah is in her car, driving through morning traffic to reach her office. She works at a marketing firm, and her days are usually busy. Her work involves analyzing data, creating marketing strategies, and coordinating teams. Her mornings are often filled with meetings and brainstorm sessions."
# sentence_original = "One ordinary day in the life of Sarah began with the soft rays of the morning sun gently nudging her awake. As she sipped on her favorite cup of steaming coffee, she pondered the possibilities that lay ahead. Sarah, a passionate photographer, embarked on a day filled with creative pursuits. Armed with her camera, she wandered through the bustling city streets, capturing candid moments and vibrant scenes that caught her discerning eye. Later, she found solace in a quiet park, where she sat beneath the shade of a sprawling oak tree, editing her photos and reflecting on the beauty she had encountered. As the sun dipped below the horizon, Sarah returned home, satisfied with the day's artistic endeavors. With a heart full of gratitude for the simple joys of life, she settled down with a good book, cherishing the tranquility that the evening brought."
```

```python
original_saic = "Sarah car morning traffic office marketing firm busy analyzing data marketing strategies coordinating teams meetings and brainstorm sessions"
original_saic = "" # supposing they analyse the sentences = frames
```

```python
import spacy
import string
import re
nlp = spacy.load('en_core_web_sm')
```

```python
def remove_space_before_punctuation(input_string):
    # Use regular expression to replace spaces before punctuation
    result = re.sub(r'\s*([.,;!?])\s*', r'\1', input_string)
    return result
```

```python
# %%timeit
sentences = [i for i in nlp(sentence_original).sents]

sentences_transmit = []
for i in range(len(sentences)):
  sentences_transmit.append(' '.join([token.text for token in sentences[i] if token.text.lower() not in set(stopwords.words('english'))]))
  sentences_transmit[-1] = remove_space_before_punctuation(sentences_transmit[-1])

# up to here, we generate the compressed sentences to transmit
```

```python
transmit = ''.join(str(x) for x in sentences_transmit)
transmit
```

```python
# calculate compression (based on the number of characters)
compression = 100 * (1 - sum(len(i) for i in transmit) / sum(len(i) for i in sentence_original))
compression
```

```python
compression_saic = 100 * (1 - sum(len(i) for i in original_saic) / sum(len(i) for i in sentence_original))
compression_saic
```

```python
from transformers import AutoTokenizer, AutoModel
from sentence_transformers import SentenceTransformer, util
```

```python
model_name_bert = 'sentence-transformers/all-mpnet-base-v2'
model_bert = SentenceTransformer(model_name_bert)
```

```python
try:
    received_gpt
    received_edge
except NameError:
  # received = "This is an important step towards equal rights for passengers"
  received_gpt = "Sarah drives her car through morning traffic to reach the office where she works at a marketing firm. Her days are usually busy, involving tasks such as analyzing data, creating marketing strategies, and coordinating teams. Mornings are often filled with meetings and brainstorm sessions."
  received_edge = "Sarah drives her car through the morning traffic to reach her office. She works at a marketing firm where her days are usually busy. Her work involves analyzing data, creating marketing strategies, and coordinating with teams. Her mornings are often filled with meetings and brainstorming sessions"

# original_saic = "important step equal rights all passengers"
reconstructed_saic = "Sarah, who works at a marketing firm, drives her car through the morning traffic to reach her office. Her busy day involves analyzing data to formulate marketing strategies, coordinating teams for various projects, and conducting meetings and brainstorm sessions to ensure effective communication and idea generation."
```

```python
received_gpt_embeddings = model_bert.encode(received_gpt, show_progress_bar=True, convert_to_tensor=True, normalize_embeddings=True)
received_edge_embeddings = model_bert.encode(received_edge, show_progress_bar=True, convert_to_tensor=True, normalize_embeddings=True)
transmit_embeddings = model_bert.encode(transmit, show_progress_bar=True, convert_to_tensor=True, normalize_embeddings=True)
original_embeddings = model_bert.encode(sentence_original, show_progress_bar=True, convert_to_tensor=True, normalize_embeddings=True)
saic = model_bert.encode(original_saic, show_progress_bar=True, convert_to_tensor=True, normalize_embeddings=True)
reconstructed_saic_embeddings = model_bert.encode(reconstructed_saic, show_progress_bar=True, convert_to_tensor=True, normalize_embeddings=True)
```

```python
print("Cosine similarity of the BERT embeddings")
print("similarity with reconstruction Chat GPT 3.5: ", 100*util.pytorch_cos_sim(original_embeddings, received_gpt_embeddings))
print("similarity with reconstruction Chat GPT 4: ", 100*util.pytorch_cos_sim(original_embeddings, received_edge_embeddings))
print("similarity without reconstruction: ", 100*util.pytorch_cos_sim(original_embeddings, transmit_embeddings))
print("improvement in comprehension: ", 100*(util.pytorch_cos_sim(original_embeddings, received_edge_embeddings) - util.pytorch_cos_sim(original_embeddings, transmit_embeddings)))
print("comparison to saic: ", 100*(util.pytorch_cos_sim(original_embeddings, received_edge_embeddings) - util.pytorch_cos_sim(original_embeddings, saic)))
print("reconstruction with saic: ", 100*(util.pytorch_cos_sim(original_embeddings, reconstructed_saic_embeddings)))
```

```python
sentences_comparison = [received_edge, transmit, original_saic]
embeddings =  model_bert.encode(sentences_comparison, show_progress_bar=True, convert_to_tensor=True, normalize_embeddings=True)
```

```python
result = util.semantic_search(original_embeddings, embeddings)


# the values are normalized, so that the choice of the most similar sentences is
# more accurate

for i in result[0]:
  print(sentences_comparison[i['corpus_id']], " - ", i)
```
